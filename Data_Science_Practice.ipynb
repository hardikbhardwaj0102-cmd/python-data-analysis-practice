```python
x=np.array([1,2,3,4,5,6])
```

```python
y=np.array([1,4,9,16,25,36])
```

```python
plt.plot(x,y)
plt.title("Squares of Numbers")
plt.ylabel("Square")
plt.xlabel("Number")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Normal")
plt.subplot(2,2,2)
plt.plot(x,x)
plt.subplot(2,2,3)
plt.plot(y,y**2)
plt.subplot(2,2,4)
plt.plot(y**2,x)
```

```python
fig=plt.figure(figsize=(6,6),dpi=100)
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2=fig.add_axes([1,1,0.8,0.8])
axis2.plot(x**2,y**2)
plt.savefig("save.png")
```
